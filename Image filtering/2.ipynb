```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import median_filter

def custom_median_filter(I, size):
    # Apply median filter
    I_median = median_filter(I, size=size)
    
    return I_median

# Load an example image
I = plt.imread("cameraman_sp.png")  # Replace "example_image.jpg" with your image file path

# Set the window size for median filtering
window_size = 3  # Change this to experiment with different window sizes

# Apply the median filter
I_median = custom_median_filter(I, window_size)

# Display the original and filtered images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(I, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')
axes[1].imshow(I_median, cmap='gray')
axes[1].set_title("Median Filtered Image (Window Size {})".format(window_size))
axes[1].axis('off')
plt.show()
```
