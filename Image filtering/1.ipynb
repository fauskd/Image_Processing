```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from skimage.color import rgb2gray

def mean_filter(I):
    # Define the mean filter kernel
    kernel = np.ones((3, 3)) / 9
    
    # Perform convolution
    I_filtered = convolve2d(I, kernel, mode='same', boundary='symm')
    
    return I_filtered

def gaussian_filter(I):
    # Define the Gaussian filter kernel
    kernel = np.array([[1, 2, 1],
                       [2, 4, 2],
                       [1, 2, 1]]) / 16
    
    # Perform convolution
    I_filtered = convolve2d(I, kernel, mode='same', boundary='symm')
    
    return I_filtered

def gradient_filter(I):
    # Define the gradient filter kernel
    kernel = np.array([[0, -1, 0],
                       [-1, 0, 1],
                       [0, 1, 0]])
    
    # Perform convolution
    I_filtered = convolve2d(I, kernel, mode='same', boundary='symm')
    
    return I_filtered

def laplacian_filter(I):
    # Define the Laplacian filter kernel
    kernel = np.array([[0, 1, 0],
                       [1, -4, 1],
                       [0, 1, 0]])
    
    # Perform convolution
    I_filtered = convolve2d(I, kernel, mode='same', boundary='symm')
    
    return I_filtered

# Load an example image
I = plt.imread("office_6.jpg")  # Replace "example_image.jpg" with your image file path

# Convert the image to grayscale
I_gray = rgb2gray(I)

# Apply the desired filter
filter_type = 'mean'  # Choose from 'mean', 'gaussian', 'gradient', 'laplacian'
if filter_type == 'mean':
    I_filtered = mean_filter(I_gray)
elif filter_type == 'gaussian':
    I_filtered = gaussian_filter(I_gray)
elif filter_type == 'gradient':
    I_filtered = gradient_filter(I_gray)
elif filter_type == 'laplacian':
    I_filtered = laplacian_filter(I_gray)
else:
    raise ValueError("Unknown filter type. Supported filters: 'mean', 'gaussian', 'gradient', 'laplacian'")

# Display the original and filtered images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(I_gray, cmap='gray')
axes[0].set_title("Original Grayscale Image")
axes[0].axis('off')
axes[1].imshow(I_filtered, cmap='gray')
axes[1].set_title("{} Filtered Image".format(filter_type.capitalize()))
axes[1].axis('off')
plt.show()
```
