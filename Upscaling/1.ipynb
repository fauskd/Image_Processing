```python
import numpy as np
import matplotlib.pyplot as plt

def resize_copy(image, factor):
    
    # Get image dimensions
    height, width, channels = image.shape

    # Create an empty array for the resized image
    new_height = height * factor
    new_width = width * factor
    resized_image = np.zeros((new_height, new_width, channels), dtype=np.uint8)

    # Copy pixels from the original image to the resized image
    for i in range(new_height):
        for j in range(new_width):
            original_i = i // factor
            original_j = j // factor
            resized_image[i, j] = image[original_i, original_j]

    return resized_image

# Load an example image
I = plt.imread("office_6.jpg")  # Replace "example_image.jpg" with your image file path

# Resize the image by a factor of 2
factor = 2
resized_image = resize_copy(I, factor)

# Display the original and resized images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(I)
axes[0].set_title("Original Image")
axes[0].axis('off')
axes[1].imshow(resized_image)
axes[1].set_title("Resized Image (Factor {})".format(factor))
axes[1].axis('off')
plt.show()
```
