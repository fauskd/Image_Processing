```python
import numpy as np
import matplotlib.pyplot as plt

#import example image
I = plt.imread('cameraman.png')

def hist_linear(I, range_min, range_max):
    min_val = np.min(I)
    max_val = np.max(I)
    scaled = (I - min_val) * (range_max - range_min) / (max_val - min_val) + range_min
    return scaled


# Apply linear histogram transformation
I_transformed = hist_linear(I, range_min, range_max)

# Define the range for pixel values
range_min = 0
range_max = 255

#plotting 
plt.imshow(I_transformed, cmap='gray', vmin=range_min, vmax=range_max)
plt.title("I_transformed")
plt.show()
```
