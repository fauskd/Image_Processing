```python
import numpy as np
import matplotlib.pyplot as plt

def hist_gamma(I, gamma):
    # Apply the gamma transformation
    I_gamma = np.power(I, gamma)
    
    # Normalize the transformed image to [0, 255]
    I_gamma = (I_gamma - np.min(I_gamma)) / (np.max(I_gamma) - np.min(I_gamma)) * 255
    
    return I_gamma
```

```python
#import image
I = plt.imread("cameraman.png")
```

Here the plot to enhance the contrast of the image

```python
#plotting for too dark

# Apply gamma transformation with different gamma values
gamma_values = [0.5, 1.0, 2.0]
plt.figure(figsize=(15, 5))
for i, gamma in enumerate(gamma_values):
    I_gamma_dark = hist_gamma(I, gamma)
    plt.subplot(1, len(gamma_values), i + 1)
    plt.imshow(I_gamma_dark, cmap='gray', vmin=0, vmax=255)
    plt.title("Gamma = {}".format(gamma))
    plt.axis('off')
plt.show()

#plotting for too bright

# Apply gamma transformation with different gamma values
plt.figure(figsize=(15, 5))
for i, gamma in enumerate(gamma_values):
    I_gamma_bright = hist_gamma(I, gamma)
    plt.subplot(1, len(gamma_values), i + 1)
    plt.imshow(I_gamma_bright, cmap='gray', vmin=0, vmax=255)
    plt.title("Gamma = {}".format(gamma))
    plt.axis('off')
plt.show()
```
