```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define the function

def hist_eq(I):
    # Compute histogram
    hist, bins = np.histogram(I.flatten(), bins=256, range=(0, 255))
    
    # Compute cumulative distribution function (CDF)
    cdf = hist.cumsum()
    
    # Normalize CDF to [0, 255]
    cdf_normalized = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())
    
    # Round and cast to integer
    cdf_normalized = cdf_normalized.astype('uint8')
    
    # Map pixels to their equalized values
    I_equalized = cdf_normalized[I]
    
    return I_equalized
```

```python
#plotting the function for the image 

# Load example image
I = plt.imread("cameraman.png")

# Convert image to integer values in [0, 255]
I_int = (I * 255).astype(np.uint8)

# Perform histogram equalization
I_equalized = hist_eq(I_int)

# Plot original and equalized images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(I_int, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(I_equalized, cmap='gray')
plt.title('Equalized Image')

plt.show()
```
