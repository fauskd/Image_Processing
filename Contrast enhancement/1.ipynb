```python
import numpy as np
import matplotlib.pyplot as plt

def my_hist(I, nbins):
    # Compute histogram manually
    min_val = np.min(I)
    max_val = np.max(I)
    bin_width = (max_val - min_val) / nbins
    
    # Initialize histogram
    H = np.zeros(nbins)
    
    # Iterate through image pixels and count occurrences in each bin
    for pixel in np.nditer(I):
        bin_index = min(int((pixel - min_val) / bin_width), nbins - 1)
        H[bin_index] += 1
    
    return H

# Sample image from the image folder
I = plt.imread('cameraman.png')


# Showing the real image
plt.imshow(I)
plt.title('Real Image')

# Number of bins
nbins =16

# Compute histogram using my_hist function
my_histogram = my_hist(I, nbins)
print('Custom Histogram:')
print(my_histogram)

# Compute histogram using NumPy's histogram function for comparison
numpy_histogram, _ = np.histogram(I, bins=nbins)
print('NumPy Histogram:')
print(numpy_histogram)

# Display image and histogram
plt.figure()
plt.imshow(I, cmap='gray', vmin=0, vmax=255)
plt.title('Original Image')
plt.colorbar()

plt.figure()
plt.bar(range(nbins), my_histogram)
plt.title('Custom Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')

plt.figure()
plt.bar(range(nbins), numpy_histogram)
plt.title('NumPy Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')

plt.show()

```
